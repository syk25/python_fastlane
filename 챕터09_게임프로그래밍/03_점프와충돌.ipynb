### 자유낙하와 바닥 충돌

스프라이트 크기 조절, 중력 구현

```python
import pygame

pygame.init()

screen = pygame.display.set_mode((1024, 768))  # 윈도우 크기
clock = pygame.time.Clock()

santa = pygame.image.load("santasprites/png/Idle (1).png").convert_alpha()
w, h = santa.get_size()
santa = pygame.transform.scale(santa, (w // 4, h // 4)) # 크기 조절
santa = santa.subsurface((50,6,86,138)) # 부분만 잘라내기
santa_rect = santa.get_rect().move(240, 20) # 산타의 시작 위치

santa_vx = 0.0 # x 방향 속도
santa_vy = 0.0 # y 방향 속도

dt = 1.0 / 30.0 # 시간 간격 (게임에서는 가급적 생략)
gravity = 1000 # 중력 가속도 (적당한 숫자로 조절)

ground_y = 512 # 바닥 위치

running = True

while running:

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    """업데이트"""
    santa_vy += gravity * dt # 속도 증가 = 가속도 * 시간 간격
    santa_rect.x += santa_vx * dt
    santa_rect.y += santa_vy * dt # 위치 이동 = 속도 * 시간 간격

    # 바닥 충돌 반응
    if santa_vy > 0 and santa_rect.bottom >= ground_y:
        santa_vy = -santa_vy * 0.5
        santa_rect.bottom = ground_y

    """화면에 그리기"""

    # 배경 색
    screen.fill((255, 255, 255))

    # 스프라이트 그리기
    pygame.draw.rect(screen, color=(128, 128, 128), rect=pygame.Rect(0, 64*8, 1024, 64*4))
    screen.blit(santa, dest = santa_rect)
    pygame.draw.rect(
        screen, color=(0, 0, 255), rect=santa_rect, width=2
    )
    pygame.display.flip()
    clock.tick(30)

pygame.quit()
```

### [참고] dt를 사용하지 않는 경우

고사양 게임에서는 곱하기 연산을 하나라도 줄이기 위해서 dt를 사용하지 않는 경우가 많습니다. dt 곱하기를 생략하는 대신 (수학적으로는 dt = 1 이라고 가정하는 것과 동일) 다른 수치들을 적당히 조절해서 원하는 움직임을 만들어냅니다. 아래 코드에서는 dt를 생략했습니다.

```python
import pygame
from pygame.locals import *

pygame.init()

screen = pygame.display.set_mode((1024, 768))  # 윈도우 크기
clock = pygame.time.Clock()

santa = pygame.image.load("santasprites/png/Idle (1).png").convert_alpha()
w, h = santa.get_size()
santa = pygame.transform.scale(santa, (w // 4, h // 4)) # 크기 조절
santa = santa.subsurface((50,6,86,140)) # 부분만 잘라내기
santa_rect = santa.get_rect().move(240, 20) # 산타의 시작 위치

santa_vx = 0.0 # x 방향 속도
santa_vy = 0.0 # y 방향 속도

#dt = 1.0 / 30.0 # 시간 간격 (게임에서는 가급적 생략)
gravity = 1 # 중력 가속도 (적당한 숫자로 조절)

ground_y = 512 # 바닥 위치

running = True

while running:

    """이벤트"""
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == KEYUP and event.key == K_SPACE:
            santa_vy = -25.0
            santa_vx = +10.0

    """업데이트"""
    santa_vy += gravity # 속도 증가 = 가속도 * 시간 간격
    santa_rect.y += santa_vy # 위치 이동 = 속도 * 시간 간격
    santa_rect.x += santa_vx

    # 바닥 충돌 반응
    if santa_vy > 0 and santa_rect.bottom > ground_y:
        santa_vx = 0 # 바닥 마찰
        santa_vy = 0 # 낙하 중지
        santa_rect.bottom = ground_y

    # 오른쪽 벽 충돌
    if santa_vx > 0 and santa_rect.right >= 1024:
        santa_vx = -santa_vx # 반사
        santa_rect.right = 1024

    """화면에 그리기"""
    screen.fill((255, 255, 255))
    screen.blit(santa, dest = santa_rect)
    pygame.draw.rect(
        screen, color=(0, 0, 255), rect=santa_rect, width=2
    )
    pygame.draw.rect(screen, color=(128, 128, 128), rect=pygame.Rect(0, ground_y, 1024, 64*4))
    pygame.display.flip()
    clock.tick(30)

pygame.quit()
```
